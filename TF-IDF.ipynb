# Definisi fungsi menghitung TF

TF adalah singkatan dari Term-Frequency. Nilai dari TF dihitung dari kemunculan kata dalam kalimat

```python
def TF(words):
    #var TF_dict untuk menampung list kata dalam kalimat yang diinputkan
    TF_dict = {}
    for word in words :
        #apabila kata sudah pernah muncul dalam TF_dict, nilai +1
        if word in TF_dict:
            TF_dict[word] + 1
        else:
            TF_dict[word] = 1
    #menghitung nilai TF dengan cara nilai kemunculan per kata dibagi dengan panjang kata dalam kalimat
    for word in TF_dict:
        TF_dict[word] = (TF_dict[word] / len(words))
    return TF_dict
```

```python
def dot_product(vector_x, vector_y):
    dot = 0.0
    for e_x, e_y in zip(vector_x, vector_y):
        dot += e_x * e_y
    return dot
```

```python
def magnitude(vector):
    mag = 0.0
    for index in vector:
        mag += math.pow(index, 2)
    return math.sqrt(mag)
```
